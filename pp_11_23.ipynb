```python
df = pd.read_csv("student_data.csv")
df.head()
df.sample(10)
```

 **Counting the students have attended at abroad or have not attended**

```python
attendedAtAbroad = df['abroad_study'].value_counts()
print(attendedAtAbroad)
```

**Filtering students who belong in "Medical" department**

```python
studentsInMedicalDept = df[df['department'] == "Medical"].sum()
print(studentsInMedicalDept)
```

No student in "Medical" department. But there is some students in "Medicine"

```python
studentsInMedicine = df[df['department'] == "Medicine"].sum()
print(studentsInMedicine)
```

**Calculating highest HSC GPA**

```python
HighestHSCGPA = df['hsc_gpa'].max()
print(HighestHSCGPA)
```

**Replacing missing values in the gender column**

```python
df.isnull().sum()
```

```python
#no missing value in the gender
# So Im putting the "unknown" in the replace of missing values in the passing_year column
replUnknwn =df.fillna({"passing_year": "Unknown"})
print(replUnknwn)
```
